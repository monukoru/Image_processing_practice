```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,10))
bricks_img = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\bricks.jpg").astype(np.float32) / 255
bricks_img = bricks_img[:,:,::-1]
bricks_gamma = np.power(bricks_img, 4)
plt.imshow(bricks_gamma)
```

```python
plt.figure(figsize=(10,10))
BLUR = cv2.blur(bricks_img,ksize=(11,11))
plt.imshow(BLUR)
```

```python
plt.figure(figsize=(10,10))
gBlur = cv2.GaussianBlur(bricks_img,ksize=(11,11),sigmaX=3)
plt.imshow(gBlur)
```

```python
plt.figure(figsize=(10,10))
mBlur = cv2.medianBlur(bricks_img,5) # it takes only odd kernal size 
plt.imshow(mBlur)
```

```python
plt.figure(figsize=(10,10))
img_sammy = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\sammy_noise.jpg")
img_denoise = cv2.medianBlur(img_sammy,5) # kernal size odd
plt.imshow(img_denoise)
```
