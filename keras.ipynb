```python
import numpy as np 
from numpy import genfromtxt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data = genfromtxt(r"C:\Users\monukoru\Downloads\image processing python\DATA\bank_note_data.txt", delimiter=',', skip_header=1)
#feature already extracted we are just working on it to train. once we learned CNN we will fetch data
labels = data[:, 4]  # here 4 is the exact number so it's not exclusive
 
features = data[:, :4]  # here 4 will be exclusve 

X = features # 2D array
y = labels  # labels 0 or 1
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

scaler_object = MinMaxScaler()
scaler_object.fit(X_train)

MinMaxScaler(copy=True, feature_range=(0,1))
scaled_X_train = scaler_object.transform(X_train)
scaled_X_test = scaler_object.transform(X_test)

model = Sequential()

model.add(Dense(4, input_dim=4, activation='relu'))

model.add(Dense(8, activation='relu'))

model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(scaled_X_train, y_train, epochs=60, verbose=2)
# epsoch is the number of times the model has gone through the data 
```

```python
model.predict_classes(scaled_X_test)
```

```python
model.metrics_names
```

```python
predections= model.predict_classes(scaled_X_test)
confusion_matrix(y_test, predections)
```

```python
print(classification_report(y_test, predections))
```
