```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cam = cv2.VideoCapture(0)

cam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)

cam.set(cv2.CAP_PROP_BUFFERSIZE, 1)  # To reduce frame in queue

face= cv2.CascadeClassifier(r"C:\Users\monukoru\Downloads\image processing python\DATA\haarcascades\haarcascade_frontalface_alt2.xml")
eye = cv2.CascadeClassifier(r"C:\Users\monukoru\Downloads\image processing python\DATA\haarcascades\haarcascade_eye.xml")

cv2.namedWindow('dectector', cv2.WINDOW_NORMAL)

while True:
    _, frame = cam.read()
    frame = cv2.flip(frame, 1)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    small_frame = cv2.resize(gray, (0,0), fx=0.5, fy=0.5)
    dimension_fc = face.detectMultiScale(small_frame, scaleFactor=1.1, minNeighbors=3, minSize=(10,10))

    for (x,y,w,h) in dimension_fc:
        x,y,w,h = x<<1, y<<1, w<<1, h<<1                                     # this is bit shifting which is fast
        if w>190:
           # eye
           roi = gray[y:y+h, x:x+w//2]
           dimension_eye = eye.detectMultiScale(roi, scaleFactor=1.1, minNeighbors=5)
           for (ex,ey,ew,eh) in dimension_eye:
               cv2.rectangle(frame, (x+ex, y+ey), (x+ex+ew, y+ey+eh), (0,255,0), thickness=2)
               break                               # this detects only one face because for loop run ones for one eye the moment it tries to go for another loop break
        else:
           cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), thickness=2)
           break          # same for the face it detects only one face 
            
    
    cv2.imshow('dectector', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cam.release()
cv2.destroyAllWindows()
```
