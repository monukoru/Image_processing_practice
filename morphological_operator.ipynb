```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
road_img = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\road white lines.jpg",0)
plt.figure(figsize=(10,10))
road_thresh = cv2.adaptiveThreshold(road_img,255 ,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV,21,9)
road_denoise = cv2.morphologyEx(road_thresh, cv2.MORPH_OPEN, np.ones((5,5), dtype=np.uint8))
plt.imshow(road_denoise, cmap='gray')
```

```python
road_img1 = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\road white lines.jpg",0)
plt.figure(figsize=(10,10))
road_thresh1 = cv2.adaptiveThreshold(road_img1,255 ,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV,21,9)
plt.imshow(road_thresh1, cmap='gray')  # original gray map
```

```python
road_img1 = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\road white lines.jpg",0)
plt.figure(figsize=(10,10))
road_thresh1 = cv2.adaptiveThreshold(road_img1,255 ,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV,21,9)
road_erode = cv2.erode(road_thresh1, np.ones((5,5), dtype=np.uint8),iterations=1)
plt.imshow(road_erode, cmap='gray')
```

```python
plt.figure(figsize=(10,10))
_,binary_thresh= cv2.threshold(road_img1, 190,255, cv2.THRESH_BINARY)
gradient = cv2.morphologyEx(binary_thresh, cv2.MORPH_GRADIENT, np.ones((4,4), dtype=np.uint8))
plt.imshow(gradient, cmap='gray')
```
