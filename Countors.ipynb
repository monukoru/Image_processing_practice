```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\internal_external.png",0)

image, contour, hierarachy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

external_countour = np.zeros(image.shape)

for i in range(len(contour)):
    # external beecause -1
    if hierarachy[0][i][3] == 4:
        cv2.drawContours(external_countour, contour, i, 255,-1)        # print and see what are the datatypes of hierarchy, cotour and image 
                                                                       # and notice that height of hierarchy and length of coutour are same 
                                                                       # for 'loop' and 'if' are just written to show the contour

plt.imshow(external_countour, cmap='gray')
```
