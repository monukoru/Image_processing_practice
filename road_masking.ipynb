```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
road_img = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\road white lines.jpg")
plt.figure(figsize=(10,10))
road_HSV = cv2.cvtColor(road_img, cv2.COLOR_BGR2HSV)
plt.imshow(road_HSV)
```

```python
# it does contain sky in it's mask
plt.figure(figsize=(10,10))
road_img_gray1 = cv2.cvtColor(road_img, cv2.COLOR_BGR2GRAY)
ret1,road_thresh1 = cv2.threshold(road_img_gray1, 170,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(road_thresh1,cmap='gray')
```

```python
# it does not contain any sky 
plt.figure(figsize=(10,10))
road_img_gray2 = cv2.cvtColor(road_img, cv2.COLOR_BGR2GRAY)
road_thresh2 = cv2.adaptiveThreshold(road_img_gray2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,21,21)
plt.imshow(road_thresh2,cmap='gray')
```
