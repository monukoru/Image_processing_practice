```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cam = cv2.VideoCapture(r"C:\Users\monukoru\Downloads\output_resized.mp4")

cv2.namedWindow('road', cv2.WINDOW_NORMAL)



while True:
    time.sleep(1/60)
    _,frame = cam.read()
    frame = cv2.cvtColor(frame)
    
    if not _:
     break
    # other piece of code
    roi = frame[frame.shape[0]//2+45:frame.shape[0], 0:frame.shape[1], :]
    
    mean_sf = cv2.pyrMeanShiftFiltering(roi, sp=20, sr=40)

    cv2.imshow('road', mean_sf)

    if cv2.waitKey(1) & 0xFF == ord('q'):
       break

cam.release()
cv2.destroyAllWindows()
```

```python
img = cv2.imread(r"C:\Users\monukoru\Pictures\image processing shots\Screenshot 2025-08-26 194129.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Mean Shift filtering (sp = spatial window, sr = color window)
mean_shift = cv2.pyrMeanShiftFiltering(img, sp=20, sr=40)

# Convert to grayscale
gray = cv2.cvtColor(mean_shift, cv2.COLOR_RGB2GRAY)

# Thresholding to get binary mask
_, segmented = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Show results
plt.figure(figsize=(12,6))
plt.subplot(1,3,1); plt.imshow(img); plt.title("Original")
plt.subplot(1,3,2); plt.imshow(mean_shift); plt.title("After Mean Shift")
plt.subplot(1,3,3); plt.imshow(segmented, cmap="gray"); plt.title("Segmented")
plt.show()
```
