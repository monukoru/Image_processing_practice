```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img =cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\sudoku.jpg",0)
plt.figure(figsize=(10,10))

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1,0,ksize=5)

sobel_y = cv2.Sobel(img, cv2.CV_64F, 0,1, ksize=5)

mix = cv2.addWeighted(src1=sobel_x,alpha=0.5,src2=sobel_y,beta=0.5,gamma=0)

gradient =cv2.morphologyEx(mix, cv2.MORPH_CLOSE, np.ones((4,4), dtype=np.uint8))
print(gradient.min())
plt.imshow(gradient, cmap='gray')
```
