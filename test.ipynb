```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image

pic = Image.open(r"DATA\00-puppy.jpg")
plt.imshow(pic)
pic_arr = np.asarray(pic).copy()  # we made a copy as direct manupulation of pixel is not possible or permitted 
pic_arr.shape
pic
```

```python
#no we will extract it's red channel just like we did in RANSAC to extract x and y values
pic_arr_RED = pic_arr[:, :, 0]
plt.imshow(pic_arr_RED, cmap='gray') #Red channel
```

```python
#no we will extract it's green channel just like we did in RANSAC to extract x and y values
pic_arr_GREEN = pic_arr[:, :, 1]
plt.imshow(pic_arr_GREEN, cmap='gray') #Red channel
```

```python
#no we will extract it's blue channel just like we did in RANSAC to extract x and y values
pic_arr_BLUE = pic_arr[:, :, 2]   #here it's shape is (width, height) and zero which is not written
plt.imshow(pic_arr_BLUE, cmap='gray') #Red channel
```

```python
#now lets extract a specific color channel (for blue)
blue_removed = pic_arr.copy()
blue_removed[:,:, 0:2] = 0
print(blue_removed.shape)
plt.imshow(blue_removed)
```

```python
#now let's extract red channel
red_removed = pic_arr.copy()
red_removed[:, :, 1:3] = 0
print(red_removed.shape)
plt.imshow(red_removed)
```

```python
#now let's do it for green
green_removed = pic_arr.copy()
green_removed[:,:, 0] = 0
green_removed[:, :, 2] = 0
print(green_removed.shape)
plt.imshow(green_removed)
```
