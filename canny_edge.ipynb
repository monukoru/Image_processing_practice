```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\sammy_face.jpg",0).astype(np.uint8)

img_blur = cv2.bilateralFilter(img, 70, 50, 10)

img_blur = cv2.morphologyEx(img_blur, cv2.MORPH_OPEN, np.ones((3,3), dtype=np.uint8))

lower = int(max(0, 0.7*np.median(img)))
upper = int(min(255,1.3*np.median(img)))

canny = cv2.Canny(img_blur, lower, upper)

plt.imshow(canny, cmap='gray')
```

```python
start = time.perf_counter()
road_img = cv2.resize(cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\road white lines.jpg",0), (640,640), interpolation=cv2.INTER_AREA)

_,road_img_thresh= cv2.threshold(road_img, 150,255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

road_img_thresh = cv2.medianBlur(road_img_thresh, ksize=(3))

road_img_thresh = cv2.erode(road_img_thresh, np.ones((2,2), dtype=np.uint8), iterations=1)
#^s4B/q5wa3;fVE4
plt.imshow(road_img_thresh, cmap='gray')

end = time.perf_counter()
print(f"time:", end-start)
```
