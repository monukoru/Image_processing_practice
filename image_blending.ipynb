```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#images of different sizes are just pasted
img2 = cv2.imread(r"DATA\watermark_no_copy.png")
img2 = img2[:,:,::-1] #watermark
img1 = cv2.imread(r"DATA\00-puppy.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) #puppy smaller
plt.figure(figsize=(10,10))

img2_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
mask = cv2.bitwise_not(img2_gray)  #now we created the mask

print(mask.shape)

masked_img = cv2.bitwise_and(img2, img2, mask=mask)

plt.imshow(masked_img)
```

```python
plt.figure(figsize=(10,10))

roi=img1[0:masked_img.shape[0], 0:masked_img.shape[1], :] 

blended = cv2.bitwise_or(roi, masked_img)

plt.imshow(blended)
```
