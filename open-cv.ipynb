```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,10))
blank_canvas = np.zeros(shape=(512, 512, 3), dtype=np.uint8)
vertices_polygon = np.array([[100,100], [300,100], [400,300], [100,200]]).reshape((-1,1,2))
cv2.polylines(blank_canvas, [vertices_polygon], isClosed=True, color=(512,512,512), thickness=1)
plt.imshow(blank_canvas)
plt.gca().invert_yaxis()
vertices_polygon
```
