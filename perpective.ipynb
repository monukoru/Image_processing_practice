```python
import numpy as np 
import cv2
```

```python
img = cv2.imread(r"C:\Users\monukoru\Downloads\perpective_image.jpg")

cv2.namedWindow('wrap', cv2.WINDOW_NORMAL)

pts = np.array([
       [894.0, 561.0],   # top-left
       [932.0, 561.0],   # top-right
       [1680.0, 933.0],  # bottom-right
       [0.0, 986.0]      # bottom-left
], dtype= np.float32) 

width ,height = 640, 480

final_pts = np.array([
       [0.0,0.0],
       [width, 0.0],
       [width, height],
       [0.0, height]
], dtype=np.float32)
 
while True:
    
 matrix = cv2.getPerspectiveTransform(pts, final_pts)

 warped = cv2.warpPerspective(img, matrix, (width, height))

 cv2.imshow('wrap', warped)

 if cv2.waitKey(1) & 0xFF == ord('q'):
  break
 
cv2.destroyAllWindows()
    

```
