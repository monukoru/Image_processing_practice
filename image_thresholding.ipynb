```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\crossword.jpg",0)
plt.figure(figsize=(20,20))

ret,thresh1 = cv2.threshold(img1, 120,214,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

plt.imshow(thresh1, cmap='gray')
```

```python
plt.figure(figsize=(20,20))
img2 = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\crossword.jpg",0)
img_thresh1 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,7,10)
plt.imshow(img_thresh1, cmap='gray')
```

```python
plt.figure(figsize=(20,20))
img2 = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\crossword.jpg",0)
ret,img_thresh2 = cv2.threshold(img2,120,255,cv2.THRESH_BINARY)
plt.imshow(img_thresh2, cmap='gray')
```

```python
plt.figure(figsize=(20,20))
blended = cv2.addWeighted(src1=img_thresh1,alpha=0.4,src2=img_thresh2,beta=0.6,gamma=0)
plt.imshow(blended,cmap='gray')
```
