```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\monukoru\Downloads\image processing python\DATA\dog_backpack.jpg").astype(dtype=np.uint8)
cv2.namedWindow('puppy', cv2.WINDOW_NORMAL)
is_drawing=  False

mask_p = np.zeros((img.shape[0], img.shape[1]), dtype=np.uint8)

def draw(event, x, y, flags, params):
    global is_drawing, mask_p

    if event == cv2.EVENT_LBUTTONDOWN:
        is_drawing = True

    elif event == cv2.EVENT_MOUSEMOVE:
        if is_drawing:
            cv2.circle(mask_p, center=(x, y), radius=15, color=255, thickness=-1)

    elif event == cv2.EVENT_LBUTTONUP:
        is_drawing = False

cv2.setMouseCallback('puppy', draw)

while True:
    masked_img = cv2.bitwise_and(img,img,mask=mask_p)
    cv2.imshow('puppy', masked_img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows()
```
