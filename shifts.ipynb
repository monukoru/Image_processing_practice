```python
import cv2
import numpy as np
import sys
```

```python
cam  = cv2.VideoCapture(0)

cv2.namedWindow('track', cv2.WINDOW_NORMAL)

_, frame = cam.read()

face = cv2.CascadeClassifier(r"C:\Users\monukoru\Downloads\image processing python\DATA\haarcascades\haarcascade_frontalface_alt2.xml")
dimension = face.detectMultiScale(frame)

(face_x, face_y, w,h) = tuple(dimension[0])   # we did tuple here because if no face is found it will return an empty array [] which will become
# an error to store in (face_x, face_y, w,h) so we tuple the dimension to safely store zero in variables

track_window = (face_x, face_y, w,h)

roi = frame[face_y:face_y+h, face_x:face_x+w]

hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

roi_hist = cv2.calcHist([hsv_roi], [0], None,[256],[0,256])

cv2.normalize(roi_hist, roi_hist, 0, 255, cv2.NORM_MINMAX)

term_crit = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 1)

while True:
    _, frame = cam.read()
    frame = cv2.flip(frame,1)
    if _ == True:  # just an optional if
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        dst =  cv2.calcBackProject([hsv], [0], roi_hist, [0, 180], 1)
        _, track_window = cv2.meanShift(dst, track_window, term_crit)
        x,y,w,h = track_window
        img = cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 3)
        cv2.imshow('track', img)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break 

cam.release()
cv2.destroyAllWindows()
```
