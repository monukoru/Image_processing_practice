```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python

img_path = r"C:\Users\monukoru\Downloads\image processing python\DATA\red_pen.jpg"
img = cv2.imread(img_path)
img = cv2.resize(img, (640, 640), interpolation=cv2.INTER_AREA)

# --- Convert only once, then split
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
draw_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # For display only

lower_red1 = np.array([0, 100, 100])
upper_red1 = np.array([10, 255, 255])

lower_red2 = np.array([160, 100, 100])
upper_red2 = np.array([180, 255, 255])

mask1 = cv2.inRange(hsv_img, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv_img, lower_red2, upper_red2)
mask = cv2.bitwise_or(mask1, mask2)

__,contour, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

largest_contour = max(contour, key=cv2.contourArea)

x,y,w,h = cv2.boundingRect(largest_contour)

cv2.rectangle(draw_img, (x,y), (x+w, y+h),(0,255,0) ,thickness=3)

plt.imshow(draw_img)
print(max(cv2.contourArea(c) for c in contour), type(__), type(contour))
```
